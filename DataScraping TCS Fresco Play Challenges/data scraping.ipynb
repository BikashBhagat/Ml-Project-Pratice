```python
import numpy as np
import pandas as pd
import bs4
from bs4 import BeautifulSoup
import requests
import re
from tqdm import tqdm
from datetime import datetime
```

```python
range_date = pd.date_range(start = '1/1/2009',end = '10/29/2018',freq = 'M')
dates = [str(i)[:4] + str(i)[5:7] for i in range_date]

len(dates)
```

```python
from tqdm import tqdm
for i in tqdm(range(20,20000000)):
  a = []
  a.append(i)
```

```python
df_list = []
index = []
for k in tqdm(range(len(dates))):
    url = "http://www.estesparkweather.net/archive_reports.php?date="+ dates[k]
    page = requests.get(url)
    page.status_code
    soup = BeautifulSoup(page.content,'html.parser')
    table = soup.find_all('table')
    parsed_data = [row.text.splitlines() for row in table]
    parsed_data =parsed_data[:-9]
    for l in range(len(parsed_data)):
        parsed_data[l] = parsed_data[l][2:len(parsed_data[l]):3]
    for i in range(len(parsed_data)):
        d = list('.'.join(re.findall("\d+",str(parsed_data[i][j].split()[:5]))) for j in range(len(parsed_data[i])))
        df_list.append(d)
        m = str(d[0])
        index.append(dates[k] + m)
    
```
