```python
#Import necessary module
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import bs4
from bs4 import BeautifulSoup
import requests
import re


# get the page from URL
url='http://marvel-ironman.surge.sh/'
page=requests.get()
page

# parse the page content is BeautifulSoup
soup = BeautifulSoup(page.content,'html.parser')

#find the content required
table = soup.findAll("div", attrs={"class": "col-lg-8 mx-auto"})
# convert xml to text and split the lines
parsed_table=[row.text.splitlines() for row in table]
# Save a separate list for Iron Man name
name = [parsed_table[i][1] for i in range(len(parsed_table))]
# save separate list for details of iron man list
detail = [parsed_table[i][3:] for i in range(len(parsed_table))]
# detail1 will contain only the description part of detail
detail1 = [[detail[i][j].split(" : ")[1] for j in range(9)] for i in range(len(detail))]
# save name in df2 data frame
df2 = pd.DataFrame(name,columns=['Ironman'])
# save name in df3 data frame
df3 = pd.DataFrame(detail1,columns=column)
# concate the data frame
df4 = pd.concat([df2,df3], axis=1)
# replace the None to np.nan
df5 = df4.replace('None', np.nan)
```
