```python
def decode_hash(hash_string):
    for i in range(1, 9):
        for j in range(33, 127):
            hash_object = hashlib.md5(str(i) + chr(j))
            guess_hash = hash_object.hexdigest()
            if guess_hash == hash_string:
                return str(i) + chr(j)
```
