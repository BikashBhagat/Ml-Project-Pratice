```python
list1 = ['Item_Fat_Content', 'Item_Type', 'Outlet_Identifier',
       'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']
fig, axs = plt.subplots(nrows=2, ncols=3,figsize = (20,10))
axs = axs.flatten()
for i in range(len(list1)):
    sns.barplot(x=list1[i] ,y = 'Item_Outlet_Sales',data=df, ax = axs[i])
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation =90)
fig.tight_layout()
```

```python
sns.pairplot(df)
```

```python
list1 = ['Item_Fat_Content', 'Outlet_Identifier',
       'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']
fig, axs = plt.subplots(nrows=2, ncols=3,figsize = (20,10))
axs = axs.flatten()

for i in range(len(list1)):
    #occurance of each feature
    feat_ocr = df[list1[i]].value_counts()
    #pie chart
    axs[i].pie(feat_ocr,labels = feat_ocr.index, startangle=90)
    axs[i].set_title(f'{list1[i]} distribution')
plt.tight_layout()
fig.delaxes(axs[5])
```

```python
num_var = ['Item_Weight', 'Item_Visibility', 'Item_MRP']
fig, axs = plt.subplots(nrows = 1,ncols=3,figsize=(20,10))
axs.flatten()

for i in range(len(num_var)):
    sns.boxplot(x=num_var[i],data = df,ax= axs[i])
plt.tight_layout()
```

```python
num_var = ['Item_Weight', 'Item_Visibility', 'Item_MRP']
fig, axs = plt.subplots(nrows = 1,ncols=3,figsize=(20,10))
axs.flatten()

for i in range(len(num_var)):
    sns.violinplot(x=num_var[i],data = df,ax= axs[i])
plt.tight_layout()
```
